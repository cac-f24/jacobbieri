```python
from xps_data_extractor import format_data
import matplotlib.pyplot as plt
from scipy.signal import peak_widths
import scipy
from scipy.ndimage import gaussian_filter1d
import scipy.signal as signal
```

```python
df = format_data('/mnt/d/Job/high resolution scan Mo, C, S, O.xlsx','high resolution scan Mo, C, S, ')
```

```python
print(df.keys())
```

```python
element = df.get("C1s")
x = element["Binding Energy (eV)"]
y = element["Intensity"]
```

```python
plt.scatter(x,y)
```

```python
smoothed_signal = gaussian_filter1d(element, sigma = 1)
```

```python
plt.plot(smoothed_signal)
```

```python
peaks, _ = signal.find_peaks(x)
print(peaks)
```

```python
bruh = scipy.signal.peak_widths(x, peaks)
```

```python
print(bruh)
```
