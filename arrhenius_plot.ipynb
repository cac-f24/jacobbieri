## Arrhenius plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
temperatures = [800,900,1000,1100]
rates = [3.24*10**-5,0.00214,0.0614,0.955]
plot_temps = []
plot_rates = []

for temp in temperatures:
    plot_temps.append(1 / temp)

for rate in rates:
    plot_rates.append(np.log(rate))
```

```python
plt.plot(plot_temps, plot_rates)
plt.title("Arrhenius plot of 1/T vs ln(k)")
plt.xlabel("1/T (1/K)")
plt.ylabel("ln(k) (1/s)")
slope, intercept, *_ = linregress(plot_temps, plot_rates)
activation_energy = 8.31 * (slope * -1)
frequency_factor = np.log(intercept)
activation_energy = f"{activation_energy:.5}"
frequency_factor = f"{frequency_factor:.5}"
print("Activation energy is:", activation_energy,"J")
print("Frequency factor is:", frequency_factor, "1/s")
```
